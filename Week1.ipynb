### 2.4.7 Combination

```python
df_group_one = accident_data[['SEVERITYCODE','WEATHER','ROADCOND','JUNCTIONTYPE','SPEEDING','ST_COLCODE']]
```

```python
grouped_test1 = df_group_one.groupby(['WEATHER','ROADCOND','JUNCTIONTYPE','SPEEDING','ST_COLCODE'],as_index=False).mean()
large = grouped_test1.nlargest(20,"SEVERITYCODE")
large
```
