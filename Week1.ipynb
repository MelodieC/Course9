# Part 3 (Modeling)

### 3.1 Predictive Modeling

We would like to apply different types of charts and regression model to determine the probability of an accident occurs.

```python
location_count  = accident_data['LOCATION'].value_counts() #count which location has the highest accident rate
location_count = location_count[:5,]
plt.figure(figsize=(200,80))
sns.barplot(location_count.index, location_count.values, alpha=1)
plt.title('The places with the highest accident rate')
plt.ylabel('Number of Accident', fontsize=72)
plt.xlabel('LOCATION', fontsize=62)
plt.xticks(fontsize=50)
plt.show()
```

```python
weather_count  = accident_data['WEATHER'].value_counts() #count under which weather has the highest accident rate
weather_count = weather_count[:5,]
plt.figure(figsize=(200,80))
sns.barplot(weather_count.index, weather_count.values, alpha=1)
plt.title('The weather condition leads to car accidents')
plt.ylabel('Number of Accident', fontsize=72)
plt.xlabel('Weather', fontsize=62)
plt.xticks(fontsize=50)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
%matplotlib inline
X = accident_data[['PEDCOUNT']]
Y = accident_data[['VEHCOUNT']]
lm = LinearRegression()
lm.fit(X,Y)
lm.score(X, Y)
```

```python
sns.regplot(x= "VEHCOUNT", y= "PEDCOUNT", data = accident_data)
```

```python
accident_data["ROADCOND"].replace("Wet", int(0), inplace = True)
accident_data["ROADCOND"].replace("Dry", int(1), inplace = True)
accident_data["ROADCOND"].replace("Unknown", int(2), inplace = True)
accident_data["ROADCOND"].replace("Snow/Slush", int(3), inplace = True)
accident_data["ROADCOND"].replace("Sand/Mud/Dirt", int(4), inplace = True)
accident_data["ROADCOND"].replace("Standing Water", int(5), inplace = True)
accident_data["ROADCOND"].replace(np.nan, int(6), inplace = True)
accident_data["ROADCOND"].replace("Ice", int(7), inplace = True)
accident_data["ROADCOND"].replace("Other", int(8), inplace = True)
accident_data["ROADCOND"].replace("Oil", int(9), inplace = True)

accident_data['ROADCOND'] = accident_data['ROADCOND'].astype(float)
accident_data[['ROADCOND','SEVERITYCODE']].corr()
#accident_data1 = accident_data.drop(accident_data.columns[0], axis=1)
#road_count1  = accident_data1['ROADCOND'].value_counts()
sns.regplot(x= accident_data["SEVERITYCODE"], y= accident_data["ROADCOND"] , data = accident_data)
```
